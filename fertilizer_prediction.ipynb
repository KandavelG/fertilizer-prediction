```python
X=df.iloc[:,0:8].values
Y=df.iloc[:,8].values
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
```

```python
pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,pred))
```

```python
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming Y_test is your true labels and pred3 is your predicted labels

# Generate classification report
report = classification_report(Y_test, pred, output_dict=True)

# Convert the report to a DataFrame for easier visualization
report_df = pd.DataFrame(report).transpose()

# Plotting the classification report
plt.figure(figsize=(10, 5))
sns.heatmap(report_df.drop(columns=['support']), annot=True, fmt=".2f", cmap="Blues")
plt.title('Classification Report')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting the comparison between actual and predicted values
plt.figure(figsize=(10, 5))
plt.plot(np.arange(len(Y_test)), Y_test, '+', label="Actual Values")
plt.plot(np.arange(len(pred)), pred, 'x', label="Predicted Values")
plt.title('Comparison between Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Class')
plt.legend()
plt.show()
```
